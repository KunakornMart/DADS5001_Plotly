```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("https://github.com/KunakornMart/DADS5001_Plotly/blob/main/Beer_FG_Feb2024.csv?raw=True")
df
```

```python
df2 = pd.DataFrame([df.groupby("Production")["Standard"].sum()])
df2=df2.T
df2.reset_index(inplace=True)
df2
```

```python
fig = px.bar(df, x="Production", y="Actual", color="Detail", title="Singha Beer Finish Cost (Baht/Litre)")

fig.add_trace(
    go.Scatter(
        mode='markers',
        x = df2['Production'],
        y = df2['Standard'],
        name="Percentage Cases",
        line=go.scatter.Line(color="gray"),
        marker_color='red',
        showlegend=False
    )
        # x=[2, 4],
        # y=[4, 8],
        # mode="lines",
        # line=go.scatter.Line(color="gray"),
        # showlegend=False)
)

fig.update_traces(
    marker=dict(size=50, symbol="line-ew", line=dict(width=4, color="Red")),
    selector=dict(mode="markers"),
)
fig.show()
```
